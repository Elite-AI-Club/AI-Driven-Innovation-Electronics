```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt

def count_faces(image_path):
    """
    Count the number of faces in an image using DeepFace

    Parameters:
    image_path (str): Path to the image file

    Returns:
    int: Number of faces detected
    list: List of face detection results with coordinates
    """
    try:
        # Read the image
        img = cv2.imread(image_path)
        if img is None:
            raise ValueError("Unable to read image file")

        # Detect faces using DeepFace
        faces = DeepFace.extract_faces(
            img_path=image_path,
            detector_backend='opencv'
        )

        # Draw rectangles around detected faces
        img_with_faces = img.copy()
        for face in faces:
            facial_area = face['facial_area']
            x = facial_area['x']
            y = facial_area['y']
            w = facial_area['w']
            h = facial_area['h']
            cv2.rectangle(
                img_with_faces,
                (x, y),
                (x + w, y + h),
                (0, 255, 0),
                2
            )

        # Convert from BGR to RGB for display
        img_with_faces_rgb = cv2.cvtColor(img_with_faces, cv2.COLOR_BGR2RGB)

        # Display the result
        plt.figure(figsize=(10, 10))
        plt.imshow(img_with_faces_rgb)
        plt.axis('off')
        plt.show()

        return len(faces), faces

    except Exception as e:
        print(f"Error processing image: {str(e)}")
        return 0, []

```

```python
image_path = "group_photo/5.jpg"
num_faces, face_data = count_faces(image_path)
print(f"Number of faces detected: {num_faces}")
```
